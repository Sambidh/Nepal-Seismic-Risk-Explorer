```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'C:/Users/Hp/Nepal Seismic Risk Explorer/nepal-earthquake-severity-index-latest.csv'
df = pd.read_csv(file_path)

# Display the first few rows of the dataset to get an overview
df.head()
```

```python
# Descriptive Statistics
descriptive_stats = df.describe()
print(descriptive_stats)
```

```python
# Visualization
sns.set(style="whitegrid")

# Histogram for Hazard Intensity
plt.figure(figsize=(10, 5))
sns.histplot(df['Hazard (Intensity)'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Hazard Intensity')
plt.xlabel('Hazard Intensity')
plt.ylabel('Frequency')
plt.show()
```

```python
# Boxplot for Severity by Poverty
plt.figure(figsize=(10, 6))
sns.boxplot(x='Severity', y='Poverty', data=df, palette='viridis')
plt.title('Boxplot of Severity by Poverty')
plt.xlabel('Severity')
plt.ylabel('Poverty')
plt.show()
```
